```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from PIL import Image

# Data
x_values = [10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000]
cpu_time = [0.001096, 0.002089, 0.003279, 0.004381, 0.004509, 0.005446, 0.006305, 0.007298, 0.008599, 0.009135, 0.010008, 0.010889]
gpu_time = [0.335073, 1.320813, 1.086546, 0.250743, 0.234915, 0.847616, 0.388539, 0.267521, 0.239413, 0.225250, 0.259564, 0.215510]

# Plot
fig, ax = plt.subplots()
ax.plot(x_values, cpu_time, color='blue', label='CPU')
ax.plot(x_values, gpu_time, color='green', label='GPU')

# Title and labels
ax.set_title('Xena CalcPi CPU v GPU performance (x10,000)')
ax.set_xlabel('Steps')
ax.set_ylabel('Time (Seconds)')

# Legend
ax.legend()

# Set x-axis ticks and labels
ax.set_xticks(x_values)
ax.set_xticklabels(x_values)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Save plot to image
canvas = FigureCanvas(fig)
canvas.draw()
image = Image.frombytes('RGB', canvas.get_width_height(), canvas.tostring_rgb())

# Save image to Photoshop file
image.save('Graph1.png')

# Show plot
plt.show()
```
